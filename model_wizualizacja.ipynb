SIR Model

```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
sp.init_printing()
```

```python
b=sp.Symbol('beta')
b
```

Opis parametrów:
S - osoby podatne,jeszcze nie zarażone </br>
I - liczba osób zakażonych </br>
Beta - współczynnik 'efektywnego kontaktu'

```python
S = sp.Function('S')
t = sp.Symbol('t')
I = sp.Function('I')
S(t),I(t)
```

```python
eq = sp.Eq(S(t).diff(t), -b*I(t)*S(t))
eq
```

```python
# Grupa 10 000 ludzi

sol = sp.dsolve(eq, S(t), ics = {S(0): 9500})
sol
# tak nie zadziała
```

```python

N = 10000
I0 = 500
S0 = N - I0 
beta= 0.2
t = np.linspace(0, 100, 100)
```

```python
# The SIR model differential equations.
def deriv(y, t, N, beta):
    S, I = y
    dSdt = -beta * S * I / N
    dIdt = beta * S * I / N # uproszczony 
    return dSdt, dIdt
```

```python
# Initial conditions vector
y0 = S0, I0
# Integrate the SIR equations over the time grid, t.
ret = odeint(deriv, y0, t, args=(N, beta))
S, I = ret.T
```

```python

fig = plt.figure(facecolor='g')
ax = fig.add_subplot(111, facecolor='#dddddd', axisbelow=True)
ax.plot(t, S,'b', alpha=0.3, lw=2, label='Podatne')
ax.plot(t, I, 'r', alpha=0.3, lw=2, label='Zarażone')
ax.set_xlabel('Dni')
ax.set_ylabel('Ilość')
legend = ax.legend()
plt.show()
```
